```python
import pandas as pd 

df = pd.read_csv("../data/raw/lat_long_kota_kab.csv")

df.head()
```

```python
location_df = df[
    [
        "name",
        "lat",
        "long"
    ]
].copy()

location_df.columns = [
    "kabupaten_kota",
    "latitude",
    "longitude"
]

location_df.head()
```

```python
# hapus kolom index lama
df = df.drop(columns=["Unnamed: 0"])

# rename kolom
df = df.rename(columns={
    "name": "kabupaten_kota",
    "lat": "latitude",
    "long": "longitude"
})

# rapikan nama wilayah
df["kabupaten_kota"] = (
    df["kabupaten_kota"]
    .str.title()
)

df.head()
```

```python
location_df = df[
    [
        "kabupaten_kota",
        "latitude",
        "longitude"
    ]
]

location_df.head()
```

```python
location_df.to_csv(
    "../data/clean/lat_long_kab_kota_clean.csv",
    index=False
)
```
