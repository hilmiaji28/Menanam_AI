```python
import pandas as pd

df1 = pd.read_csv("../data/raw/Crop_Recommendation_Non Cassava.csv")
df2 = pd.read_csv("../data/raw/Crop_recommendation.csv")
```

```python
#rename datset 1 untuk crop
df1 = df1.rename(
    columns={"label": "crop_label"}
)
```

```python
#kolom soil ditambahkan untuk dataset 1 (unknown)
df1["soil"] = "unknown"
```

```python
columns = [
    "N",
    "P",
    "K",
    "temperature",
    "humidity",
    "ph",
    "rainfall",
    "soil",
    "crop_label"
]

df1 = df1[columns]
df2 = df2[columns]
```

```python
#gabungkan dataset 1 dan dataset 2
combined_df = pd.concat(
    [df1, df2],
    ignore_index=True
)
```

```python
#hapus duplikasi 
print("Duplicate:", combined_df.duplicated().sum())

combined_df = combined_df.drop_duplicates()
```

```python
#cek hasil
print(combined_df.shape)

print(combined_df.head())

print(combined_df["crop_label"].value_counts())
```

```python
#dataset baru (final-gabungan)
combined_df.to_csv(
    "../data/clean/crop_recommendation_combined.csv",
    index=False
)
```
