```python
#korelasi 
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.select_dtypes(
    include="number"
).corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    cmap="coolwarm"
)

plt.show()
```
